```python
import tensorflow as tf
import sounddevice as sd
import soundfile as sf
import librosa
import IPython.display as ipd
import numpy as np
from matplotlib import pyplot as plt
```

```python
label = "test"
Directory = "../Data/AudioFragments/"
HZ = 16000
# HZ = 8000
duration = 1  # seconds
```

```python
filename = label + '.wav'
print("start")
mydata = sd.rec(int(HZ * duration), samplerate=HZ,
                channels=1, blocking=True)
print("end")
sd.wait()
sf.write(Directory + filename, mydata, HZ)
```

```python
#reading the voice commands
samples, sample_rate = librosa.load(Directory + filename, sr=HZ)
samples = librosa.resample(samples, sample_rate, HZ)
ipd.Audio(samples, rate=HZ)
```
